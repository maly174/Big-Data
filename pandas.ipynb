```python
import pandas as pd
import numpy as np

# Parametry
start_date = '2024-01-01 12:00'
frequency = '15min'
data_start = 10
data_end = 20

# Tworzenie danych
values = np.arange(data_start, data_end)
num_periods = len(values)  # Automatyczne określenie liczby okresów

# Tworzenie serii danych
dates = pd.date_range(start=start_date, periods=num_periods, freq=frequency)
series = pd.Series(data=values, index=dates)

print(series)
```
